```python
import numpy as np
from hydra import initialize, compose
import json

with initialize(version_base=None, config_path="../../src/conf/"):
    cfg = compose(config_name='keyword_extraction.yaml', overrides=["local=default"])
```

```python

import os

#project = 'Waikato|weka-3.8-903-04804ccd6dff03534cbf3f2a71a35c73eef24fe8.json'
project = 'GenomicParisCentre|eoulsan-5391-0fac2ac7fa5da73c794824de1dbf216099ebefc2.json'
#project = 'SonarSource|sonar-java-8464-6400749499be832a3e37fa2f6beed47f47c04f36.json'
#project = 'PolaricServer|aprsd-169-1f180c80a586fbcb301acc2526715bd6386a86ba.json'
project_name = project.split("|")[0]
method = 'name'
path = os.path.join(cfg.annotations_path,'kl', method, project)

with open(f"/home/sasce/PycharmProjects/CodeGraphClassification/data/processed/annotations/{method}/label_mapping.json", 'rt') as inf:
    label_map = json.load(inf)
```

```python
import json

with open(path, 'rt') as inf:
    annot = json.load(inf)

files = list(annot.keys())
annotations = []
for k in files:
    annotations.append(annot[k])
```

```python
import plotly.express as px
annotations = np.array(annotations)
fig = px.imshow(annotations, text_auto=True, aspect="auto",  width=800, height=400) #, color_continuous_scale=[[0, 'white'], [0.5, 'rgb(92, 228, 255)'], [1.0, 'rgb(0, 0, 255)']])
fig.update_layout({
    'plot_bgcolor': 'rgba(0,0,0,0)',
    'paper_bgcolor': 'rgba(0,0,0,0)'
})
fig.update_layout(margin=dict(l=1, r=1, t=1, b=1), font_color="black")
fig.write_image(f"/home/sasce/PycharmProjects/CodeGraphClassification/reports/plots/annotations/{project_name}_{method}_label_headmap.pdf")
fig.show()
```
