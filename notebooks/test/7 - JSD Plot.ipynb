```python
from scipy.spatial.distance import jensenshannon
import numpy as np
```

```python
N = 100
uniform = np.ones(N)
uniform = uniform / sum(uniform)
```

```python
uniform
```

```python
high = np.zeros(N)
high = np.random.uniform(0, .1, N)
high[25] = 2
high[5] = 3
high[85] = 9
high = high / sum(high)
```

```python
import matplotlib.pyplot as plt

x = list(range(0, N))
plt.axis([0, N-1, 0, .2])
plt.plot(x, high, uniform)
```

```python
jensenshannon(high, uniform)
```

```python
low = np.random.uniform(0, N, N)
low = low /sum(low)
```

```python
import matplotlib.pyplot as plt
plt.axis([0, N-1, 0, .2])
x = list(range(0, N))
plt.plot(x, low, uniform)
```

```python
jensenshannon(uniform, low)
```

```python
low
```

```python
import csv
with open('resources/example_jsd.csv', 'wt') as outf:
    writer = csv.writer(outf)
    writer.writerow(['funct', 'x'])
    for x, y in enumerate(high):
        for i in range(int(y*10000)):
            writer.writerow(['High', x])
    for x, y in enumerate(low):
        for i in range(int(y*10000)):
            writer.writerow(['Low', x])
    for x, y in enumerate(uniform):
        for i in range(int(y*10000)):
            writer.writerow(['Uniform', x])
```
